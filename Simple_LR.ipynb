```python
#G1------------G3

data=df[['G1','G3']]


#Input --------------X
X=data['G1']

#Output--------------y
y=data['G3']
```

```python
#Create a figure 

plt.figure(figsize =(8,8))

sns.scatterplot(data=data,x='G1', y='G3', color ='orange')
plt.title('First Period Grade---End of Course Grade')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Lr=LinearRegression()
```

```python
X.shape
```

```python
X=X.values.reshape(-1,1)
```

```python
y.shape
```

```python
y=y.values.reshape(-1,1)
```

```python
#Let's import train test split from sklearn

#70 % --- Train Data 

#30 % --- Test Data

#X_train, y_train === Train_Data
#X_test, y_test === Test_Data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3, random_state=100)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
#Let's fit Train Data

Lr.fit(X_train,y_train)
```

```python
#Let's calculate intercept

Lr.intercept_
```

```python
#Let's calculate coef

Lr.coef_
```

```python
#y=B0+B1X

#B0=-1.82
#B1=1.11

#y=-1.82+1.11X
```

```python
y_pred=Lr.predict(X_test)
```

```python
y_pred[0:10]
```

```python
y_test[0:10]
```

```python
y_pred.shape
```

```python
y_test.shape
```

```python
#We should create a figure to show real and prediction data 

# Original label -- y_test

#Prediction label -- y_pred

# Input --X_test
```

```python
# Original Data ---Grand Truth 

fig,ax = plt.subplots(figsize =(8,8))
ax.scatter(X_test,y_test, label = 'Grand Truth', color ='orange')

# Prediction Data 

plt.plot(X_test,y_pred, label ='Prediction', color ='red')
         

    
plt.title('First Period Grade---End of Course Grade')
plt.xlabel('First Period Grade')
plt.ylabel('End of Course Grade')
plt.legend(loc='upper left')
plt.show()
```

```python
#Let's see changes in each prediction point
indexs =range(1,120)

#Original Data---- Grand Truth
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(indexs, y_test, label ='Grand Truth', color ='red', linewidth =2)

# Prediction

ax.plot(indexs, y_pred, label = 'Prediction', color ='green', linewidth =2)

plt.title('Grand Truth--Prediction')
plt.xlabel('Data Index')
plt.ylabel('End of Course Grade')
plt.legend(loc='upper left')
plt.show()
```

```python
#let's find residuals on the figure (y_test-y_pred)
```

```python
import numpy as np
```

```python
#Residuals
indexs =range(1,120)

#Original Data---- Grand Truth
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(indexs, y_test-y_pred, label ='Residuals', color ='orange', linewidth =2)

# Prediction

ax.plot(indexs, np.zeros(119), color ='green', linewidth =2)

plt.title('Error--Residuals')
plt.xlabel('Data Index')
plt.ylabel('End of Course Grade')
plt.legend(loc='upper left')
plt.show()
```

```python
#Let's control the accuracy of the model
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
#Let's calculate R^2 score 

R_2=r2_score(y_test,y_pred)
R_2
```

```python
R_2 * 100
```

```python
#R squared = 58.71 %

# Our model represents 58.71 % of real data.
```

```python
#Let's calculate mean squared error
mse =mean_squared_error(y_test,y_pred)
mse
```

```python
#Let's calculate RMSE 
import math 

RMSE=math.sqrt(mse)
RMSE
```

```python
# RMSE = 2.92

# Our model makes an average of 2.92 errors.
```
