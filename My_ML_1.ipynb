```python
#Let's find the relationship between studytime and G1.
data=dataset[['studytime','G1']]

#Input--->X

X=data['studytime']

#Output--->Y

y=data['G1']


#Average Absences --->y_mean

y_mean = data.groupby('studytime').mean()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, color ='orange')

plt.plot(y_mean.index,y_mean, color ='red')

plt.title('Studytime--G1')
plt.xlabel('Studytime')
plt.ylabel('G1')
plt.show()
```

```python
#Let's find the relationship between school and G1.

data1=dataset[['school','G1']]


#Input-----X1

X1=data1['school']

#Output ------y

y=data1['G1']


#Average y_mean

y_mean=data1.groupby('school')
```

```python
#Create a figure 

y_mean.boxplot(subplots=False, figsize =(8,8))


plt.title('School--G1')
plt.xlabel('School')
plt.ylabel('G1')


plt.show()
```

```python
#Let's find the relationship between  Absences and  G1 .
data2=dataset[['absences','G1']]


#Input--->X

X=data2['absences']

#Output--->y

y=data2['G1']

#Average of absences 

y_mean=data2.groupby('absences').mean()
```

```python
# Creata a figure 
fig,ax= plt.subplots(figsize =(8,8))
ax.scatter(X,y, color ='blue')

ax.plot(y_mean.index,y_mean, color ='red')

plt.title('Absences---G1')
plt.xlabel('Absences')
plt.ylabel('G1')


plt.show()
```
